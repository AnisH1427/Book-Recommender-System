### Popularity Based Recommender System

```python
rating_with_bookName = ratings.merge(books,on='ISBN')
```

```python
rating_with_bookName.head()
```

```python
num_rating = rating_with_bookName.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
num_rating.tail()
```

```python
avg_rating = rating_with_bookName.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating.tail()
```

```python
popular_df = num_rating.merge(avg_rating,on='Book-Title')
```

```python
popular_df.head()
```

```python
#getting top 50 books whoose number of rating is greater than 250
popular_df = popular_df.query('num_ratings >= 250').sort_values('avg_ratings', ascending=False).head(50)
popular_df
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

```python
popular_df.head()
```

```python
popular_df.describe()
```

## Collaboraative filtering Based Recommender System

```python
#filtering user which has over 200 ratings
x=rating_with_bookName.groupby('User-ID').count()['Book-Rating']>200
selected_user = x[x].index
selected_user[:5]
```

```python
#filtering data of selected user only
filtered_rating = rating_with_bookName[rating_with_bookName['User-ID'].isin(selected_user)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.head()
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt.info()
```

```python
pt.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    #fetch index from book name
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar_items:
        item=[]
#         print(pt.index[i[0]])
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        #using extend inplace of append to avoid higher dimensional list
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    return data
```
